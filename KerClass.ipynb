```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='diagnosis', data=df)
```

```python
def map_diagnosis(diagnosis):
    if diagnosis == 'B':
        return 0
    elif diagnosis == 'M':
        return 1
    else:
        return None  # Handle other cases if necessary

# Applying the mapping function to create the new column
df['diagnosisN'] = df['diagnosis'].apply(map_diagnosis)
```

```python
df.head()
```

```python
df.drop('diagnosis', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.corr()['diagnosisN'].sort_values().plot(kind='bar')
```

```python
X = df.drop('diagnosisN', axis=1).values
y = df['diagnosisN'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

model.add(Dense(1,activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#Callbacks in the case of overfitting
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
mod_losses = pd.DataFrame(model.history.history)
```

```python
#dropout layers for turning off some neurons
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid')) #binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test,y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history).plot()
```

```python
predict_x=model.predict(X_test) 
classes_x=np.argmax(predict_x,axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, classes_x))
```

```python
print(classification_report(y_test, classes_x))
```
